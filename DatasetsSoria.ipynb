## Data set 1: Camapaña de Marketing

El dataset trata de campañas de marketing de una empresa y sus resultados, contiene además, datos personales de los clientes como ingresos, composición familiar, nivel de ingresos, numero de compras, respuestas a campañas de marketing, entre otros.


```python
marketing = pd.read_csv("Marketing_campaign.csv", sep="\t")
```

```python
marketing.info()
```

```python
marketing["Dt_Customer"] = pd.to_datetime(marketing["Dt_Customer"], format='%d-%m-%Y', dayfirst=True)
```

```python
marketing.info()
```

```python
marketing["Income"].fillna(marketing["Income"].mean(), inplace=True)
```

```python
marketing.info()
```

```python
marketing
```

## Data set 3: Satisfaccion de clientes de una aerolinea
El dataset trata de la satisfacción de cada cliente con una aerolínea, conteniendo datos personales del cliente, el tipo de viaje, la distancia, le evaluación del cliente ante diferentes variables, entre otros.

```python
airline = pd.read_csv("Airline_customer_satisfaction.csv", sep=",")
```

```python
airline.info()
```

```python
airline["Arrival Delay in Minutes"].fillna(airline["Arrival Delay in Minutes"].mean(), inplace=True)
```
